```python
import pandas as pd
need_to_load = False

if need_to_load:
    df = pd.read_csv("cruelty_free.csv", engine = "python")
    df = df[1:]
    df.columns = ['Company Name', 'Exchange', 'Ticker', 'Explanation']
```

```python
import yfinance as yf
import numpy as np
need_to_download = False

if need_to_download:
    subset1 = df[df['Exchange'] == 'NASDAQ']
    names1 = subset1['Ticker'].tolist() # 2148 stocks!
    names1 = np.unique(names1)
    names1_str =  ' '.join(names1)
    prices1 = yf.download(tickers=names1_str, period='5y', interval='1d', actions=False) #340 failed downloads
```

```python
if need_to_download:
    subset2 = df[df['Exchange'] == 'NYSE']
    names2 = subset2['Ticker'].tolist() # 1943 stocks!
    names2 = np.unique(names2)
    names2_str =  ' '.join(names2)
    prices2 = yf.download(tickers=names2_str, period='5y', interval='1d', actions=False) #285 failed downloads
```

```python
if need_to_download:
    prices1.to_csv("cruelty_free_prices1.csv")
    prices2.to_csv("cruelty_free_prices2.csv")
```
